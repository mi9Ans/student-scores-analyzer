```python
import pandas as pd

data = {
    'Name' : ['Asha', 'Ravi', 'Zara', 'Pooja', 'Aman' ],
    'Math' : [85, 70, 95, 35, 50],
    'Science' : [90, 75, 92, 38, 60],
    'English' : [88, 68, 96, 32, 45]
}

df=pd.DataFrame(data)
df.head()

class Student :
  def __init__(self, name, scores):
    self.name = name
    self.scores = scores

  def average(self):
    return sum(self.scores.values()) / len(self.scores)

  def has_passed(self):
    return self.average() >= 40

students = []

for i in range(len(df)):
  name = df.iloc[i]['Name']
  scores = {
      'Math':df.iloc[i]['Math'],
      'Science':df.iloc[i]['Science'],
      'English':df.iloc[i]['English']
  }
  s = Student(name, scores)
  students.append(s)

def print_all_averages(students):
  for s in students:
    print(f"{s.name}'s average is {s.average(): .2f}")

def get_topper(students):
  return max(students, key=lambda s: s.average())

print_all_averages(students)
topper = get_topper(students)
print(f"\n Topper is {topper.name} with average {topper.average(): .2f}")

import matplotlib.pyplot as plt

top3 = sorted(students, key=lambda s: s.average(), reverse=True)[:3]

names = [s.name for s in top3]
avgs = [s.average() for s in top3]

plt.bar(names, avgs, color='skyblue')
plt.title("Top 3 Students")
plt.ylabel("Average Score")
plt.show()

```
